**factorial design**

- the experiment where there are one or more factorial

- the joint effect has to be considered
- and
- one or more respoonse variable.

**2^k factorial design**

- multiple factors
  - each factor only has two levels/treatment.

- assumption
  - randomly assignment
  - the assumptions for error

- usually used in begining of experiment
  - to reduce the number of factor to study.
 

**2^2 factorial design**

- A
  - two levels
- B
  - two levels

- ex)
  - each combination of levels from each factors
    - has 3 replicates.


- main effect
  - change in response
  - based on
    - change in levels in one factor averaged by the levels of other factors.
   

- in the case when the two factors are A and B
  - levels of each factors are two

- the design can be written as rectangle.
  - ab - (A is high level/ B is high level)
  - a - (A is high level/ B is low level)
  - b - (A is low level / B is high level)
  - (1) - ( A is low level / B is low level)

- and each combination has n replications.


-  the main effect of A
  - the difference of averages 
    - yA+
    - and
    - yA-
    - = (yA+ - yA-)
  - yA+ = (ab + a) / 2n
  - yA- = (b + (1) ) / 2n
  - so the main effect of A i

- the main effect of B
  - the difference of averages
    - yB+
    - and
    - yB-
    - = (yB+ - yB-)
   
  - yB+ = (ab +  b) / 2n
  - yB- = (a + (1)) / 2n

- the main effect of interaction AB
    - average difference of
       - yAB+
       - and
       - yAB-
      - or
       - yA+B
       - and
       - yA-B
      - = (yAB+ - yAB-) / 2
      - or
      - = (yA+B - YA-B ) /2
      - yAB+ = (ab - b) / n
      - yAB- = (a - (1)) / n
      - yA+B = (ab - a ) / n
      - yA-B =(b - (1) ) / n

- another way of getting the main effect of interaction
  - the difference between the average of interaction and not interaction
  - yab = ( ab + (1)) /2n

  -  so main effect of interaction is
      - yab - yno
