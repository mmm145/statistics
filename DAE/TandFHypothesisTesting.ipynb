**case**

- dependent variable
  - categorical
    - factor 
- independent variable
  - numerical

**model**

- model
  - y(i,j) =  mu(i) + e(i,j)
    - i = 1,2,... number of factor
    - j = 1,2,... number of observation
    - e
      - from N(0, var( i ) )
      - independent with each other.
  - u(i)s are constant, so you can think the following as equivalent
    - y(i,j)
      - from N( mu(i), var(i) )
- model assumption
  - independence
    - satisfied with randomized 
      - run order
      - experimental unit
  - constant variance
    - check with normal probability plot
  - normality
    - check with normal probability plot 


**hypothesis**

- null hypothesis
  - mu(i) are the same
- alternative hypothesis
  - at least one mu is different
  - especially when i = 2
  - mu(1) != mu(2)

**steps**

- take random sample
- set the hypothesis
- compute
  - test statistics
  - p-value
- reject of fail to reject the hypothesis.

**decision making( two sided )**

- check the value of alpha(type one error)
  - probability
    - rejecting the null hypothesis
    - based on distribution where null hypothesis is true.
- if two sided,
  - you want to make one sided alpha / 2 

**test statistics**

- t distribution
  - degree of freedom
  - n(i) + n(j) - 2
- under the assumption of equal variance << can be tested by the F-test. 
  - ( sample mean(i) - sample mean(j) ) / standard error
    - common standard variance (sp2) 
      - (# of i - 1 ) * ( sample variance of i ) + ( # of j - 1 ) *( sample variance of j )
      - divided by
      - (# of i + # of j -2)
    - standard error
      - sp * sqrt( (1 / # of i) + (1 / # of j) )
     

      

**two possible errors**

- type 1 error
  - alpha
  - error when
    - reject null hypothesis
    - when
    - null hypothesis is true. 
- type 2 error
  - beta
  - error when
    - fail to reject null hypothesis
    - when
    - null hypothesis is not true
   
- power
  - 1- beta
  - power to reject the null hypothesis when the null hypothesis is not true

**confidence interval**

- value from sample (+/-) t(alpha / 2) * sp * (normalize based on if population variance are equal or not)

- the comparision between hypothesis testing and confidence interval

    - null hypothesis is mu(1) = mu(2)
    - hypothesis test
      - use p-value
      - reject the null hypothesis if the p-value is less than alpha.
    - confidence interval
      - if the confidence interval
      - does not include
      - 0
      - then mu(1) != mu(2)
      - with 1 - alpha probability.
     
- the margin of error
  - ( upper bound - lower bound ) / 2
  - the smaller, the more precise the interval estimation is.
- especially, t - test for comparision of mean
  - the margin error
  - t( alpha / 2 )(degree of freedom)* sp * sqrt( 1/ n1 + 1/n2 )
    - =>
    - the larger the sample size, the better.
    - the larger the sample size, the more power ( reject the null hypothesis when null hypothesis is not true)
