# one way anova

**model 1**

- y(i,j) = mu(i) + e(i,j)
  - y(i,j) - the value of observation.
  - mu(i) - the mean of the treatment i.
  - e(i,j) - the error of i,j th observation.
    - follow N(0, var^2)
    - all var should be the same regardless of the levels in a factor


**model 2**

- y(i,j) = mu + t(i) + e(i,j)
  - y(i,j) - the value of observation. 
  - mu - the overall mean
  - t(i) - the effect of treatment i
  - e(i,j) - the error for each observation.
    - follows N(0, var^2)
    - all var should be the same regardless of the levels in a factor. 



# assumptions for anvoa.

- e(i,j) are independent
  - run order vs residual 
- e(i,j) are normally distributed.
  - QQ plot
- var^2 are the same regardless of treatment
- var^2 are constant

# summary


- point estimate
  - y(i) - y(j)
- standard error
  - sqrt(2 * MSE / n )
 
- confidence interval for mu(i) - mu(j)
  - (point estimate) +- (critical value) * (standard error)

- hypothesis testing
  - null hypothesis
    - mu(i) == mu(j)
  - alternative hypothesis
    - mu(i) != mu(j)
 - test statistics
 - estimate / standard error
 - check the vlaue of
   - abs(test statistics)
   - and
   - critical value.
