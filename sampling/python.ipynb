```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# basic (for linear regression) 

```python
n_samples = 1000
variance_noise = 10
w1 = 3
w0 = 10
np.random.seed(1)

x = 10 * np.random.rand(n_samples)
y = w1 * x + w0 + np.random.normal(0, variance_noise,n_samples)
# linear relationship
# constant variance 
# normal 
# 
y_hat = w1 * x + w0
```

```python
plt.scatter(x,y)
plt.plot(x,y_hat,color = "red")
```
